```python
import pandas as pd
import numpy as np
import os


def add_date(filename):
    df = pd.read_csv(filename)
    a=filename
    sep = '.,-'
    for i in sep:
        file_name_split = (a.split('.'))
        file_name_without_ext = (file_name_split[0])
        spliting_filename = (file_name_without_ext.split('-'))
    date = spliting_filename[-1]
    if(date[-8] == 0):
        year = date[-4: ]
        month = date[2:4]
        day = date[:2]
        date_of_purchase = year + '-' + month + '-' + day
        df['date_of_purchase'] = date_of_purchase
        df.to_csv(filename)
        return;
```

```python
success = []
failed = []

s=os.listdir('/home/bala/oddanchatram/data/dindigul/2014')
for i in s:
    try:
        add_date(i)
        success.append(i)
    except:
        failed.append(i)
```
